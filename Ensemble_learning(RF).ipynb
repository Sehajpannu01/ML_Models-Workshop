Ensemble Learning: Combining multiple models to create a single model. This will help in making single strong models which will create positive effect on output.

We'll use multiple-decision tree to predict values, also will use different algos and combine them to create strong models.
> Techniques used:
- Bagging 
- Boosting
- Stacking (Voting)

>Ensemble (group) learning is a method of ML that 
  improves accuracy by combining models.

>It means, multiple weak models are combined to 
  represents a strong model


The three most popular methods for combining the 
predictions from different models are:

1.Bagging:
--------------
 
>Building multiple models (typically of the same type) 
from different subsamples of the training dataset.

	Commonly used Bagging techniques are:
	-------------------------------------
	>RandomForestClassifier,RandomForestRegressor


2.Boosting: 
---------------

>Building multiple models (typically of the same type) each 
of which learns to fix the prediction errors of a prior model 
in the chain.
	
	Commonly used Boosting techniques are:
	-------------------------------------
	>AdaBoostClassifier,AdaBoostRegressor
	>GradientBoostClassifier,GradientBoostRegressor

3.(Stacking)Voting: 

>Building multiple models (typically of differing types) and
simple statistics (like calculating the mean) are used to 
combine predictions.
 
	Commonly used Voting techniques are:
	-------------------------------------
	>VotingClassifier
	>VotingRegressor
