```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from matplotlib import style
from sklearn.preprocessing import StandardScaler
import seaborn as sb
import warnings
from word2number import w2n 
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_csv("Assets/Height_Weight_Dataset.csv")
df
```

```python
X = df.iloc[:,0].values
```

```python
X
```

```python
X.shape #need to convert to 2d array before passing it to model
```

```python
X= X.reshape(-1,1)
```

```python
X.shape
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state=0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
pred = lr.predict(X_test)
```

```python
r2_score(y_test, pred) 
#See accuracy is little less let's see if this data is linear or non-linear by ploting
```

```python
style.use('seaborn-v0_8-deep')
plt.scatter(X,y, color = 'Black', label = 'Data Points')
plt.plot(X,y, color = 'pink',label = 'Regression Line')
plt.xlabel('Age')
plt.ylabel('Height')
plt.show()
```
