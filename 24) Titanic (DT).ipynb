### Let's drop unwanted columns from training dataframe

```python
X = df.drop(columns = 'Survived', axis= 1)
X.head()
```

```python
X = X.drop(columns = ['PassengerId','Ticket','Name'], axis= 1)
X.head()
```

### Let's convert object into integer so that pass to the algorithm

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler
```

```python
lb = LabelEncoder()
```

```python
X['Sex'] = lb.fit_transform(X['Sex'])
```

```python
X.head()
```

```python
X['Embarked'] = lb.fit_transform(X['Embarked'])
X.head()
```

```python
X['Age'] = X['Age'].astype(int)
X.head()
```

```python
sd = StandardScaler()
X = sd.fit_transform(X)
```

```python
X
```

```python
y = df['Survived']
y.head()
```

```python
y.shape
```

```python
X.shape
```

### Let's split data into testing and training

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3, random_state = 42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier,plot_tree
```

```python
dt = DecisionTreeClassifier(criterion='entropy',random_state=42,max_depth=4)
```

```python
dt.fit(X_train,y_train)
```

```python
pred = dt.predict(X_test)
pred
```

```python
# Let's evaluate
from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)
```

```python
features = ['Pclass','Sex','Age','SibSp','Parch','Fare','Embarked']
```

```python
# Let's plot decision tree
import matplotlib.pyplot as plt
```

```python


plt.figure(figsize=(25,30))
plot_tree(dt, feature_names=features,class_names='Survived', filled=True, fontsize=10)

plt.show()
```
