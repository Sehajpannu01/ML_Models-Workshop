```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from matplotlib import style
from sklearn.preprocessing import StandardScaler
import seaborn as sb
import warnings
from word2number import w2n 
```

```python
df = pd.read_csv('Assets/hiring.csv')
df.head()
```

```python
df.shape
```

# Need to clean this data first

```python
#Lets convert data of experience in numerical form using a liberary called word2number:
```

```python
pip install word2number
```

```python
df['experience'] 
```

```python
df['experience'] = df['experience'].apply(lambda x:w2n.word_to_num(x) if isinstance(x,str) else x) 
```

```python
df
```

```python
df.shape
```

```python
df['experience'].mean()
```

```python
df.loc[[1],'experience'] = df.loc[[1],'experience']=5
df
```

```python
df.loc[[0],'experience'] = df.loc[[0],'experience'].fillna(4)
df
```

```python
df.loc[[6],'test_score(out of 10)'] = df.loc[[6],'test_score(out of 10)'].fillna(7)
df
```

```python
df.drop('test_score(out of 10)\t',axis = 1, inplace = True)
```

```python
df
```

# Finally we have clean data

```python
#Now let's scale the data and split it
```

```python
X = df.iloc[:,:-1]
```

```python
X
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(X)
```

```python
X
```

```python
X.shape
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = .2, random_state = 0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.coef_  #three coef means multi-linear regression is applied automatically
```

```python
lr.intercept_
```

```python
pred = lr.predict(X_test)
```

```python
pred
```

```python
r2_score(y_test, pred) #due to less data model is underfit
```
