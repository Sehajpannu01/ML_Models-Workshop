```python
# Drop existing Age column from train and test if it exists
if 'Age' in train.columns:
    train.drop(columns=['Age'], inplace=True)

# Merge Age column from df into train and test dataframes
train_1 = train.merge(df[['PassengerId', 'Age']], on='PassengerId', how='left')
```

```python
columnstoremove = ['PassengerId', 'Survived','Ticket','Cabin']

train_1.drop(columns= columnstoremove, axis= 1, inplace= True)

train_1.drop(columns= ['Name'], axis= 1, inplace= True)
```

```python
train_1.head()
```

```python
train_1.isna().sum()
```

```python
# Fill NaN values in the Embarked column with the mode

mode_embarked = train_1['Embarked'].mode()[0]
train_1['Embarked'] = train_1['Embarked'].fillna(mode_embarked)
```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

lb = LabelEncoder()
train_1['Sex'] = lb.fit_transform(train_1['Sex'])
train_1['Embarked'] = lb.fit_transform(train_1['Embarked'])

s_c = StandardScaler()
train_1 = s_c.fit_transform(train_1)
```

```python
train_1
```

```python
target = train['Survived']
target
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(n_estimators=10,random_state=0, criterion='entropy')
rf.fit(train_1, target)
```
