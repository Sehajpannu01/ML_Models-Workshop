```python
# Drop existing Age column from train and test if it exists
if 'Age' in train.columns:
    train.drop(columns=['Age'], inplace=True)

# Merge Age column from df into train and test dataframes
train_1 = train.merge(df[['PassengerId', 'Age']], on='PassengerId', how='left')


```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')

test['Age'] = imputer.fit_transform(test[['Age']])
test['Fare'] = imputer.fit_transform(test[['Fare']])

```

```python
from sklearn.preprocessing import LabelEncoder, StandardScaler

lb = LabelEncoder()
train_1['Sex'] = lb.fit_transform(train_1['Sex'])
train_1['Embarked'] = lb.fit_transform(train_1['Embarked'])

test['Sex'] = lb.fit_transform(test['Sex'])
test['Embarked'] = lb.fit_transform(test['Embarked'])

s_c = StandardScaler()
train_1 = s_c.fit_transform(train_1)
test = s_c.fit_transform(test)

```

```python
target = train['Survived']
target
```

```python
test
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, BaggingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score,f1_score, confusion_matrix
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(train_1, target, test_size=0.2, random_state=42)

classifiers = [
    RandomForestClassifier(n_estimators= 5,random_state=42),
    AdaBoostClassifier(random_state=42),
    BaggingClassifier(random_state=42),
    LogisticRegression(max_iter=1000, random_state=42),
    KNeighborsClassifier(),
    DecisionTreeClassifier(random_state=42)
]
```

```python
results = []

for clf in classifiers:
    clf_name = clf.__class__.__name__
    print(f'\nEvaluating Model {clf_name}-->')
    clf.fit(X_train, y_train)

    pred = clf.predict(X_test)
    accuracy = accuracy_score(y_true=y_test, y_pred=pred)
    precision = precision_score(y_true=y_test, y_pred=pred)
    recall = recall_score(y_true=y_test, y_pred=pred)
    f1 = f1_score(y_true=y_test, y_pred=pred)

    print(f"Accuracy: {accuracy:.2f}")
    print(f"Precision: {precision:.2f}")
    print(f"Recall: {recall:.2f}")
    print(f"F1 Score: {f1:.2f}")

```

```python
rf_model = RandomForestClassifier(n_estimators= 5,random_state=42)
rf_model.fit(X_train, y_train)

pred_1 =  rf_model.predict(test)
```

```python
test_write_to_excel = pd.read_csv('Assets/titanic/test.csv')
# test_write_to_excel.head()
passenger_ids = test_write_to_excel['PassengerId']
```
