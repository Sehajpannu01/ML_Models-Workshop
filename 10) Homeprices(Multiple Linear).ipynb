```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from matplotlib import style
from sklearn.preprocessing import StandardScaler
import seaborn as sb
import warnings
```

```python
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Assets/homeprices (1).csv")
df.head()
```

```python
average_value= df['bedrooms'].mean()
average_value
```

```python
df.fillna(4.0, inplace = True) #filled average value in NaN. Cannot remove because data is very small.
```

```python
df
```

```python
numerical = ["area","bedrooms","age"]
```

```python
ss = StandardScaler()
```

```python
X = ss.fit_transform(df[numerical])
```

```python
X
```

```python
X.shape
```

```python
y = df.iloc[:,-1]
```

```python
y
```

```python
y.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = 0)
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
pred= lr.predict(X_test)
```

```python
r2_score(y_test, pred) #Model is underfit due to very less data
```

## Let's train model again on whole data

```python
lr.fit(X,y) #trained model with whole data
```

```python
pred= lr.predict(X_test)
```

```python
r2_score(y_test, pred) 
```
