```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score
from matplotlib import style
import warnings
```

```python
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv("Assets/homeprices_area.csv")
```

```python
df
```

```python
X = df.iloc[:,:-1].values
```

```python
X.ndim
```

```python
X
```

```python
y = df['price']
```

```python
y.ndim
```

```python
lr = LinearRegression()
```

```python
lr.fit(X,y)
```

```python
predict_df = pd.read_csv('Assets/areas.csv')
```

```python
predict_df
```

```python
predict_df.drop('price', inplace = True, axis = 1)
```

```python
predict_df.shape
```

```python
pred = lr.predict(predict_df)
```

```python
pred
```

```python
predict_df['Price'] = pred 
pd.set_option('display.float_format', '{:.2f}'.format)
```

```python
predict_df
```

```python
predict_df.to_csv('Assets/areas.csv', index = False)
```
